**BMI** (Body Mass Index) is a measure to assess body weight in relation to height. It’s calculated using the formula:

BMI = Weight(Kg) / Height(m)^2

\

According to the World Health Organization (WHO) standards, the BMI categories are as follows:

**Underweight:** less than 18.5

**Normal weight:** 18.5 - 24.9

**Overweight:** 25 - 29.9

**Obese:** 30 or higher
